```julia
using CSV
using DataFrames
using Dates
using Plots

files = readdir("D:\\TC_Kirrily\\")

num = 1
for ii in files

    infil = "D:\\TC_Kirrily\\"* ii

    global wind_df = DataFrame(CSV.File(infil, header=20, footerskip=2));
    site_name = string(wind_df[1,"wmo"])*" "*wind_df[1,"name[80]"]

    wind_df.Date = DateTime.(string.(wind_df[:,"local_date_time_full[80]"]), "yyyymmddHHMMSS");

    if num == 1
        global p1 = plot(wind_df.Date,wind_df.wind_spd_kmh, lw=:2, label=site_name)
    else
        p1 = plot!(wind_df.Date,wind_df.wind_spd_kmh, lw=:2, label=site_name)
    end

    num += 1
   
end

p1_plot = plot(p1, size=(1200,800), framestyle = :box,fg_legend=:transparent, bg_legend=:transparent, legend=:topleft,
        leftmargin = 15Plots.mm, grid=true, gridlinewidth=0.5, gridstyle=:dot, gridalpha=1, title="Wind speed")

savefig(".\\townsville_wind_data_plot.png")

display(p1_plot)
```

```julia
num = 1
for ii in files

    infil = "D:\\TC_Kirrily\\"* ii

    global wind_df = DataFrame(CSV.File(infil, header=20, footerskip=2));
    site_name = string(wind_df[1,"wmo"])*" "*wind_df[1,"name[80]"]

    wind_df.Date = DateTime.(string.(wind_df[:,"local_date_time_full[80]"]), "yyyymmddHHMMSS");
    
    sectors = ["N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE", "S", "SSW", "SW", "WSW", "W", "WNW", "NW", "NNW"]

    bearings = zeros(nrow(wind_df))
    for ii in 1:length(sectors)

        bearings[findall(dir .== sectors[ii])] .= 22.5 * ii - 22.5

    end

    if num == 1
        global p1 = plot(wind_df.Date,bearings, lw=:3, label=site_name)
    else
        p1 = plot!(wind_df.Date,bearings, lw=:3, label=site_name)
    end

    num += 1
   
end

p1_plot = plot(p1, size=(1200,800), ylim=(0,360), yflip = true, framestyle = :box,fg_legend=:transparent, bg_legend=:transparent, legend=:bottomright,
        leftmargin = 15Plots.mm, grid=true, gridlinewidth=0.5, gridstyle=:dot, gridalpha=1, title="Wind Direction")

savefig(".\\townsville_direction_data_plot.png")

display(p1_plot)
```
